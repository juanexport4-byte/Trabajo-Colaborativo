# Introducción al Aprendizaje Automático (Machine Learning)

El **Machine Learning (ML)** o **Aprendizaje Automático** es una rama de la Inteligencia Artificial (IA) que permite a las computadoras aprender y tomar decisiones a partir de los datos, sin necesidad de ser programadas explícitamente para cada escenario.

En la programación tradicional, un humano escribe un conjunto de reglas (instrucciones) y le da datos a la computadora para obtener un resultado. En **Machine Learning**, el enfoque se invierte: se le dan los datos y los resultados esperados a la computadora, y ella utiliza algoritmos matemáticos para **descubrir el patrón o la regla** por sí misma.

---

## 💡 ¿Cómo aprende una máquina?

El proceso de aprendizaje se divide principalmente en tres grandes metodologías, dependiendo de la naturaleza de los datos disponibles:

### 1. Aprendizaje Supervisado (El enfoque de este repositorio)
El algoritmo se entrena utilizando un conjunto de datos que ya incluye las "respuestas correctas" (etiquetas). El objetivo es aprender la función que conecta las entradas con las salidas para predecir futuros casos.
*   **Ejemplos:** Predicción del precio de una vivienda, clasificación de correos en "Spam" o "No Spam", y modelos de **Regresión Lineal**.

### 2. Aprendizaje No Supervisado
Aquí los datos no tienen etiquetas ni respuestas predefinidas. El algoritmo debe explorar la estructura de la información por su cuenta para encontrar patrones ocultos, agrupaciones o similitudes.
*   **Ejemplos:** Segmentación de clientes de un banco por comportamiento de compra o sistemas de recomendación de películas.

### 3. Aprendizaje por Refuerzo
El algoritmo (llamado agente) aprende a base de interactuar con un entorno dinámico mediante un sistema de **premios y castigos**. Su objetivo es maximizar la recompensa a largo plazo a través del ensayo y error.
*   **Ejemplos:** Software que aprende a jugar ajedrez o sistemas de navegación para autos autónomos.
