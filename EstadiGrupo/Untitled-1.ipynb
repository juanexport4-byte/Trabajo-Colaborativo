# Evaluación de Modelos: Métricas de Rendimiento en Regresión



Una vez que un modelo de Machine Learning ha sido entrenado, no podemos asumir que sus predicciones son correctas. Necesitamos cuantificar su rendimiento. Este módulo explica las métricas fundamentales para evaluar algoritmos de regresión y presenta un ejemplo práctico de cálculo.



---



## 1. Métricas Fundamentales



Para medir la distancia entre los valores reales ($y$) y las predicciones del modelo ($\hat{y}$), se utilizan principalmente tres métricas estadísticas:



| Métrica | Definición Matemática | Propósito Práctico |

| :--- | :--- | :--- |

| **MAE** *(Error Absoluto Medio)* | $$\frac{1}{n} \sum_{i=1}^{n} \vert y_i - \hat{y}_i \vert$$ | Mide el promedio de los errores absolutos. Es fácil de interpretar porque mantiene las mismas unidades que la variable original. |

| **MSE** *(Error Cuadrático Medio)* | $$\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$ | Eleva los errores al cuadrado antes de promediarlos. Penaliza fuertemente las predicciones que se alejan mucho del valor real (valores atípicos). |

| **R²** *(Coeficiente de Determinación)* | $$1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$ | Mide la proporción de la varianza de los datos que es explicada por el modelo. Varía entre 0 y 1 (donde 1 es un ajuste perfecto). |



---



## 2. Ejemplo Práctico de Cálculo



Imaginemos que entrenamos nuestro modelo de Machine Learning para predecir el rendimiento de un cultivo (en toneladas) basándonos en el uso de fertilizantes. Tras evaluar **4 parcelas de prueba**, obtenemos los siguientes datos:



* **Valores Reales ($y$):** `[10, 12, 15, 18]`

* **Predicciones del Modelo ($\hat{y}$):** `[11, 14, 14, 16]`



### Tabla de Desglose de Errores



Para calcular el MAE y el MSE, construimos la siguiente relación paso a paso:



| Dato ($i$) | Valor Real ($y$) | Predicción ($\hat{y}$) | Error Simple ($y - \hat{y}$) | Error Absoluto ($\vert y - \hat{y} \vert$) | Error al Cuadrado ($(y - \hat{y})^2$) |

| :---: | :---: | :---: | :---: | :---: | :---: |

| 1 | 10 | 11 | -1 | 1 | 1 |

| 2 | 12 | 14 | -2 | 2 | 4 |

| 3 | 15 | 14 | 1 | 1 | 1 |

| 4 | 18 | 16 | 2 | 2 | 4 |

| **Suma** | - | - | - | **6** | **10** |



### Cálculo de Resultados:



1. **Cálculo del MAE:**

Dividimos la suma de los errores absolutos entre el número de datos ($n = 4$):

$$MAE = \frac{6}{4} = 1.5\text{ toneladas}$$

*Interpretación:* En promedio, las predicciones del modelo se equivocan por 1.5 toneladas.



2. **Cálculo del MSE:**

Dividimos la suma de los errores al cuadrado entre el número de datos ($n = 4$):

$$MSE = \frac{10}{4} = 2.5$$

## 3. El Fenómeno del Overfitting (Sobreajuste)



Al evaluar métricas, el Compañero 4 debe recordar una regla de oro en Machine Learning: **un error de cero en el entrenamiento no siempre es bueno**.



* **Overfitting:** Ocurre cuando el modelo memoriza el ruido de los datos de entrenamiento de forma tan perfecta que sus errores son casi cero. Sin embargo, al probarlo con datos nuevos del mundo real, el error se dispara porque el modelo perdió la capacidad de generalizar.
