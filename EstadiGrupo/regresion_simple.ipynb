# Fundamentos Matemáticos de la Regresión Lineal Simple

La **Regresión Lineal Simple** es uno de los algoritmos fundacionales y más utilizados en el **Machine Learning Supervisado**. Su objetivo principal es modelar y entender la relación funcional entre una variable independiente (predictora) y una variable dependiente (criterio o respuesta).

---

## 1. La Ecuación del Modelo

En el ámbito teórico, asumimos que la relación entre las variables sigue una tendencia lineal interceptada por un componente aleatorio de error. La ecuación formal se define como:

$$y = \beta_0 + \beta_1 x + \epsilon$$

Donde cada término representa lo siguiente:
*   $y$: Variable dependiente o respuesta (lo que queremos predecir).
*   $x$: Variable independiente o predictora (el dato que conocemos).
*   $\beta_0$ (Intercepto): El valor esperado de $y$ cuando $x = 0$. Es el punto donde la recta cruza el eje vertical.
*   $\beta_1$ (Pendiente): El cambio estimado en $y$ por cada unidad que incrementa $x$. Define la inclinación y dirección de la recta.
*   $\epsilon$ (Error aleatorio o residuo): Captura la variabilidad que el modelo lineal no puede explicar (ruido, variables omitidas, etc.).

---

## 2. Estimación por Mínimos Cuadrados Ordinarios (MCO)

Dado que en el mundo real no conocemos los verdaderos parámetros $\beta_0$ y $\beta_1$, el algoritmo de Machine Learning utiliza un conjunto de datos de entrenamiento para calcular sus estimaciones, denotadas como $\hat{\beta}_0$ y $\hat{\beta}_1$. La ecuación de predicción resultante es:

$$\hat{y} = \hat{\beta}_0 + \hat{\beta}_1 x$$

Para encontrar la "recta de mejor ajuste", el método más común es el de **Mínimos Cuadrados Ordinarios (MCO)**. Este enfoque busca minimizar la suma de los errores al cuadrado (la distancia vertical entre los puntos reales del dataset y la línea recta que dibuja el modelo).

### Función de Pérdida (Error Cuadrático Medio - ECM)
El algoritmo aprende minimizando la siguiente función:

$$ECM = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Al derivar esta función respecto a $\beta_0$ y $\beta_1$ e igualar a cero, se obtienen las fórmulas analíticas definitivas para los estimadores:

$$\hat{\beta}_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2} = \frac{Cov(X, Y)}{Var(X)}$$

$$\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}$$

*(Donde $\bar{x}$ y $\bar{y}$ representan las medias aritméticas de las variables $x$ e $y$ respectivamente).*

---

## 3. Supuestos Críticos del Modelo

Para que las inferencias y predicciones de este modelo de Machine Learning sean matemáticamente válidas y confiables, los datos deben cumplir con cuatro supuestos fundamentales (conocidos frecuentemente bajo las siglas **LINE**):

1.  **Linealidad:** La relación entre la variable independiente $x$ y la media de la variable dependiente $y$ debe ser lineal.
2.  **Independencia:** Las observaciones y sus respectivos errores $\epsilon$ deben ser independientes entre sí (no debe haber autocorrelación).
3.  **Normalidad:** Para cualquier valor fijo de $x$, los errores $\epsilon$ deben distribuirse siguiendo una distribución normal con media cero.
4.  **Homocedasticidad:** La varianza de los errores debe ser constante a lo largo de todos los niveles de la variable predictora $x$ (no debe abrirse en forma de abanico).
